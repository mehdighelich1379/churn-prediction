```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
data = pd.read_csv('../data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
print(data.nunique())
```

```python
data.drop(columns=['EmployeeCount','Over18','StandardHours','EmployeeNumber'] , inplace=True)
```

```python
data.describe()
```

```python
sns.countplot(data=data , x='Attrition' , hue='Attrition')
plt.show()
```

#### Do younger employees with lower income tend to leave more?

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=data, 
    x='Age', 
    y='MonthlyIncome', 
    hue='Attrition',  
    alpha=0.6
)
plt.title('Age vs Monthly Income by Attrition')
plt.show()
```

## ðŸ“Œ Insight from Plot â€“ Age vs Monthly Income by Attrition

The scatter plot shows that employees with a monthly income below $10,000 are more likely to leave the company.

 This trend is especially pronounced among younger employees under the age of 35, suggesting a higher attrition risk in this group despite lower salaries.

#### Do employees with high overall experience but low tenure in the current company tend to leave?

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=data ,
    x='TotalWorkingYears' ,
    y='YearsAtCompany'  ,
    hue='Attrition',  
    alpha=0.6)
plt.title('TotalWorkingYears vs YearsAtCompany')
plt.legend()
plt.show()
```

## ðŸ“Œ Insight from Plot â€“ Total Working Years vs Years at Company

The majority of attrition cases are observed among employees with less than 20 total working years and under 15 years of tenure at the current company. 

This suggests that employees in the earlier to mid stages of their careers are more likely to leave, possibly due to seeking better opportunities or dissatisfaction in early roles.

#### Are employees with long commutes and low income more likely to leave?

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=data ,
    x='DistanceFromHome' ,
    y='MonthlyIncome'  ,
    hue='Attrition',  
    alpha=0.6)
plt.title('Attrition vs DistanceFromHome and MonthlyIncome')
plt.legend()
plt.show()
```

## ðŸ“Œ Insight from Plot â€“ Attrition vs Distance From Home and Monthly Income

The scatter plot indicates that employees with a monthly income below $10,000 and who commute between 5 and 25 kilometers are more likely to leave the company.
Notably, there is a visible concentration of attrition cases in the 15â€“20 km range, suggesting that moderate commute distances combined with lower salaries might contribute to higher turnover.

#### Are those who haven't been promoted for a long time more likely to leave?

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=data ,
    x='YearsSinceLastPromotion' ,
    y='YearsAtCompany'  ,
    hue='Attrition',  
    alpha=0.6)
plt.title('YearsSinceLastPromotion vs YearsAtCompany')
plt.legend()
plt.show()
```

## ðŸ“Œ Insight from Plot â€“ Years Since Last Promotion vs Years at Company

The plot suggests that attrition is more common among employees who have not been promoted for several years, especially those who have been with the company for a short to moderate period (under ~15 years).
A noticeable number of churn cases appear when employees have not received a promotion in over 5 years, indicating a potential link between lack of career growth and employee turnover.

Younger employees (below ~35) show a noticeably higher attrition rate compared to older employees.

#### Attrition by Job Role (Bar Plot)


```python
plt.figure(figsize=(20, 10))
sns.countplot(data=data, x='JobRole', hue='Attrition')
plt.show()
```

Attrition rates vary significantly by job role.
Sales Representatives, Laboratory Technicians, and Human Resources roles have the highest proportion of attrition, while roles such as Managers and Research Directors exhibit the lowest turnover.





####  Monthly Income vs Attrition (Box Plot)

```python
sns.boxplot(data=data, x='Attrition', y='MonthlyIncome')
plt.show()
```

Employees who left the company generally had lower median monthly incomes compared to those who stayed.
The box plot clearly shows that attrition is concentrated among lower-income employees.

#### Distance from Home vs Attrition (Violin Plot)

```python
sns.violinplot(data=data, x='Attrition', y='DistanceFromHome')
```

Employees who left tend to have slightly higher median commuting distances.
The density plot shows two peaks in attrition: one for very short distances and another for employees living further away (~15â€“25 km).

```python
sns.pairplot(data, vars=['Age', 'MonthlyIncome', 'TotalWorkingYears'], hue='Attrition')
```

####  Attrition Rate by Job Satisfaction (Crosstab)

```python
pd.crosstab(data['JobSatisfaction'], data['Attrition'], normalize='index')
```

Job satisfaction is strongly associated with attrition:

Employees with the lowest job satisfaction (1) have the highest attrition rate (22.8%).

Attrition decreases as job satisfaction increases, dropping to 11.3% for employees with the highest satisfaction (4).

#### Department vs Attrition (Crosstab)

```python
pd.crosstab(data['Department'], data['Attrition'], normalize='index')*100
```

Employees in the Sales department have the highest attrition rate (20.6%), followed by Human Resources (19%).
The Research & Development department has the lowest attrition rate at 13.8%.

####  OverTime vs Attrition (Crosstab)

```python
pd.crosstab(data['OverTime'], data['Attrition'], normalize='index')
```

Employees who work overtime have a much higher attrition rate (30.5%) compared to those who do not (10.4%).
This suggests that workload and work-life balance may be key drivers of turnover.

####  Marital Status vs Attrition (Crosstab)

```python
pd.crosstab(data['MaritalStatus'], data['Attrition'], normalize='index')
```

Single employees have the highest attrition rate (25.5%), which is significantly higher compared to Married (12.4%) and Divorced employees (10%).

#### Monthly Income by JobRole and Attrition (Boxplot)

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=data, x='JobRole', y='MonthlyIncome', hue='Attrition')
plt.xticks(rotation=45)
plt.show()
```

Across most job roles, employees who left tend to have lower median monthly income compared to those who stayed.
Roles like Sales Executive and Laboratory Technician show the biggest income gaps between attrition groups.

####  Training Times Last Year vs Attrition (Crosstab)

```python
pd.crosstab(data['TrainingTimesLastYear'], data['Attrition'], normalize='index')*100
```

Employees who received no training last year have the highest attrition rate (27.8%).
As the number of training sessions increases, the attrition rate decreases, reaching only 9.2% for employees with 6 training sessions.

#### Total Working Years (Binned) vs Attrition (Crosstab)

```python
data['WorkingYearBin'] = pd.cut(data['TotalWorkingYears'], bins=[0,5,10,15,20,40], 
                                labels=['0-5','6-10','11-15','16-20','20+'])

pd.crosstab(data['WorkingYearBin'], data['Attrition'], normalize='index')*100

```

Employees with 0-5 total working years have the highest attrition rate (28%).
Attrition rate gradually decreases as total experience increases, reaching only 7.7% for employees with 20+ years of experience.

####  Pairplot (Age, Monthly Income, Distance from Home)

```python
sns.pairplot(data, vars=['Age','MonthlyIncome','DistanceFromHome'], hue='Attrition')
```

Attrition is more concentrated among younger employees, those with lower monthly income, and employees with short or moderate commuting distances.
The pairplot highlights clear clusters where younger, lower-income employees are more likely to leave.

#### Years Since Last Promotion vs Attrition Rate (Line Plot)

```python

data['Attrition_num'] = data['Attrition'].map({'Yes': 1, 'No': 0})


data.groupby('YearsSinceLastPromotion')['Attrition_num'].mean().plot(kind='line', marker='o')

plt.ylabel('Attrition Rate')
plt.title('Attrition Rate by Years Since Last Promotion')
plt.show()
```

Attrition tends to be higher among employees who have not received a promotion for a long time, particularly after 6-8 years without promotion.
There is significant fluctuation, but the general trend suggests that lack of promotion can contribute to turnover.

#### Work-Life Balance + OverTime vs Attrition

```python
pd.crosstab([data['WorkLifeBalance'], data['OverTime']], 
            data['Attrition'], normalize='index')*100
```

Attrition is significantly higher among employees with poor work-life balance (1) who also work overtime. For example, employees with Work-Life Balance = 1 and working overtime have an attrition rate as high as 45%, compared to only 25% for those who do not work overtime.

#### Environment Satisfaction + Performance Rating vs Attrition

```python
pd.crosstab([data['EnvironmentSatisfaction'], data['PerformanceRating']], 
            data['Attrition'], normalize='index')*100
```

Employees with low environment satisfaction (1 or 2) and low-to-average performance ratings (3) have the highest attrition rate (up to 26%). This suggests dissatisfaction with the work environment combined with moderate performance ratings may increase turnover.
